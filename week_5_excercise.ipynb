# Stage One: Import Libraries / Data

I imported the pandas library and created a dataframe object from my Avengers Endgame Tweets. 

```python
import pandas as pd

avengers_df = pd.read_csv('Avengers4tweets.csv', delimiter=",")

avengers_df
```

# Stage Two: Display a Summary and Sub-sections of the Data

First, I pulled a random sample of 10 tweets in order to get a general idea of the types of tweets in the data set. 

I then pulled the most common 10 accounts that users replied to in order to understand which accounts users were replying to most. 

Lastly, I ran a search in the text of the tweets for the keyword "woman" in order to see how much users discussed the women (characters, actors, producers) involved in the movie (note: some of my results contain "woman" inside usernames, and the word "women" is not included in this dataset). 

```python
avengers_df.sample(10)

avengers_df['replyToSN'].value_counts()[:10]

avengers_df[avengers_df['text'].str.contains('woman')]
```

# Stage Three: Clean Your Data

I had a lot of sections in my dataset that read NaN, particularly for the values that showed the usernames and tweet ID of the original post a user replied to. I replaced those values with "not a reply."  

I also dropped duplicates in the text column in order to trim out tweets that had been retweeted multiple times.


```python
avengers_df['replyToSN'] = avengers_df['replyToSN'].fillna('Not a reply')
avengers_df['replyToSID'] = avengers_df['replyToSID'].fillna('Not a reply')
avengers_df['replyToUID'] = avengers_df['replyToUID'].fillna('Not a reply')
avengers_df = avengers_df.drop_duplicates('text') 

avengers_df
```

# Stage Four: Plot Your Data

First, I used the screenName value to determine the most active users by the number of tweets they authored and created a pie chart. 

Then, I created a bar graph showing which accounts were being replied to most to show what accounts users were responding to. To take out the most common "not a reply" I collected values 1 - 16. 


Lastly, I created a line graph based on when tweeting was the most common, by tranfroming the "created" column to show minutes instead of seconds and plotting a graph based on that value. Of note, his type of visualation would be more helpful in a data set taken over a longer period of time. 


```python
avengers_df['screenName'].value_counts()[:15].plot(kind="pie", title="Most Active Users")
```

```python
avengers_df['replyToSN'].value_counts()[1:16].plot(kind="barh", title="Most Replied to Accounts")
```

```python
def truncate_to_minutes(created):
    return created[:16]

minutes_df = avengers_df['created'].transform(truncate_to_minutes)
minutes_df.value_counts().plot(kind="line", title="Tweets over Time")

```
